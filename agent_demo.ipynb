### Using DuckDuckGo search tool

```python
from langchain.agents import initialize_agent, Tool
from langchain_openai import ChatOpenAI
```

```python
from langchain_community.tools import DuckDuckGoSearchRun
```

```python
search_tool = Tool(
    
    name = "DuckDuckGo Search",
    func = DuckDuckGoSearchRun().run,
    description = "Use this tool to perform web search using DuckDuckGo"
)
```

```python
llm = ChatOpenAI(model="gpt-4.1-mini", temperature = 0.7)

tool = [search_tool]

agent = initialize_agent(tool, llm, agent ="zero-shot-react-description", verbose = True, handle_parsing_errors=True)
response = agent.run("tell me Best Movies of 2025?")
print(response)
```
